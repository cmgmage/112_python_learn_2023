```python
import numpy as np
import pandas as pd
scores = np.random.randint(20,100,size=(50,5))
scores_df = pd.DataFrame(scores,
             columns = ['國文','英文','數學','地理','歷史'],
             index = range(1,51))
scores_df
```

```python
scores_df['總分'] = scores_df['國文'] + scores_df['英文'] + scores_df['數學'] + scores_df['歷史'] + scores_df['地理']
scores_df['平均'] = scores_df['總分'] / 5
scores_df
```
