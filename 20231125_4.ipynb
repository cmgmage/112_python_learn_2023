```python
import numpy as np
import pandas as pd
scores = np.random.randint(20,100,size=(50,5))
scores_df = pd.DataFrame(scores,
             columns = ['國文','英文','數學','地理','歷史'],
             index = range(1,51))
display(scores_df)

sum_values = scores_df.sum(axis=1)
mean_values = scores_df.mean(axis=1)
scores_df['總分'] = sum_values
scores_df['平均'] = mean_values
scores_df
```

```python
scores_df.sort_values(by='總分')
```

```python
#前五名
scores_df.sort_values(by='總分',ascending=False).iloc[:5]
```

```python
#後五名
scores_df.sort_values(by='總分',ascending=True).iloc[:5]
```

```python
#使用subscript[]，過濾資料
scores_df[scores_df['平均'] < 80]
```

```python
#平均80以下，歷史不及格的
scores_df[(scores_df['平均'] < 80) & (scores_df['歷史'] < 60)]
```

```python
#使用query()的搜尋
scores_df.query('平均<80 and 歷史<60') 
```
